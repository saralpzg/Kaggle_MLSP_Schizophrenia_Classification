# Introducción a los datos

* **Características FNC:**

FNC = Functional Network Connectivity. Las variables de este conjunto se refieren a valores de correlación en la conexión en mapas del cerebro (tomadas mediante MRI), independientes en el tiempo. Por lo que al final con FNC feature obtenemos una imagen de los patrones de conexiones funcionales en el tiempo. Las imágenes de MRI en cada momento indicaban el nivel de actividad en la zona.

* **Características BSM:**

SBM = Source-Based Morphometry. Estas variables recogen los pesos asociados a los mapas obtenidos al aplicar ICA (independent component analysis) en las concentraciones de materia gris de todos los sujetos. Si toma un valor cercano a 0 para un mapa cerebral derivado de ICA, esto indica que esas regiones del cerebro del mapa no están muy presentes, es decir, hay poca concentración de materia gris.

**t-SNE** (T-distributed Stochastic Neighbor Embedding) es un algoritmo diseñado para la visualización de conjuntos de datos de alta dimensionalidad. Si el número de dimensiones es muy alto, Scikit-Learn recomienda en su documentación utilizar un método de reducción de dimensionalidad previo (como PCA) para reducir el conjunto de datos a un número de dimensiones razonable (por ejemplo 50), lo que reducirá el ruido y aligerará la ejecución de t-SNE.

t-SNE se ejecuta en dos pasos: en primer lugar construye una distribución de probabilidad sobre parejas de muestras en el espacio original, de forma tal que las muestras semejantes reciben alta probabilidad de ser escogidas, mientras que las muestras muy diferentes reciben baja probabilidad de ser escogidas. El concepto de "semejanza" se basa en la distancia entre puntos y densidad en las proximidades de un punto.

En segundo lugar, t-SNE lleva los puntos del espacio de alta dimensionalidad al espacio de baja dimencionalidad de forma aleatoria, define una distribución de probabilidad semejante a la vista en el espacio destino (el espacio de baja dimensionalidad), y minimiza la denominada divergencia Kullback-Leibler entre las dos distribuciones con respecto a las posiciones de los puntos en el mapa (la divergencia de Kullback-Leibler mide la similitud o diferencia entre dos funciones de distribución de probabilidad). Dicho con otras palabras: t-SNE intenta reproducir la distribución que existía en el espacio original en el espacio final.

Algoritmo en sklearn:

* n_components (2 por defecto): Dimensiones del conjunto transformado.
* perplexity (30 por defecto): Según la documentación de Scikit-Learn, se recomienda un valor entre 5 y 50 (mayor cuanto mayor sea el dataset), aunque se indica que el algoritmo no es muy sensible a este valor.
* early_exaggeration (12 por defecto): Este parámetro controla la distancia entre bloques semejantes en el espacio final. La elección de este valor no es crítico.
* learning_rate (200 por defecto): Habitualmente en el rango (10-1000). Si es muy elevado, los datos transformados estarán formados por una bola de puntos equidistantes unos de otros. Si es muy bajo, los puntos se mostrarán comprimidos en una densa nube con algunos outliers.
* n_iter (1000 por defecto): Número máximo de iteraciones para la optimización. Debería ser, por lo menos, 250.
* metric: métrica para la medición de las distancias.
* method: algoritmo a usar para el cálculo del gradiente.

# Outliers en train

El problema de los datos outliers en esta clasificación es especialmente importante, ya que el volumen de los datos disponibles para el entrenamiento es muy reducido (86 pacientes) y por tanto, los modelos serán más sensibles a la presencia de datos anómalos. 

Se considerarán outliers todas aquellas muestras cuyos datos sean consistentemente anómalos. Sin embargo, como tampoco se desea podar excesivamente un dataset ya bastante reducido, la exigencia será alta y el criterio adoptado para tratar de "consistentemente anómalo" un registro es exigir un 80% de sus datos como outliers (por debajo o por encima de los percentiles 10 y 90 de la variable respectivamente).

```python
percent10, percent90 = [], []

for col in train_data.columns[1:]:
    serie = pd.Series(train_data[col])
    percent10.append(np.percentile(serie, 10))
    percent90.append(np.percentile(serie, 90))

threshold = round(train_data.shape[1]*0.8) # Al menos 80% de los valores de la fila (paciente) son sospechosos
print("Si un paciente presenta más de {} datos anómalos, sus datos se considerarán 'outliers'.".format(threshold))
outliers = [] # Guarda los IDs de los pacientes consistentemente anómalos
for idx, row in train_data.iterrows():
    num_outliers = (row[1:] < percent10).sum()
    if num_outliers > threshold:
        outliers.append(row[0])
        
outliers
```

Además, vamos a comprobar que las muestras de ambos conjuntos están balanceadas, es decir, ninguna etiquetas tiene un volumen de muestras muy superior al volumen de la contraria.

```python
control_group = train_labels[train_labels["Class"]==0]
sick = train_labels[train_labels["Class"]==1]
plt.bar(["Grupo de control", "Pacientes de esquizofrenia"], [len(control_group), len(sick)], width=0.2)
plt.show()
```
